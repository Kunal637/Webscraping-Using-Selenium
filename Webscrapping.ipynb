```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
import re

# Set up the WebDriver for Microsoft Edge
driver = webdriver.Edge()

# Define the search query
query = "watch"

# Navigate to the Flipkart search page for watches
driver.get(f"https://www.flipkart.com/search?q={query}watch&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")

# Wait for the page to load
time.sleep(3)

# Find all product items based on their class name
elems = driver.find_elements(By.CLASS_NAME, 'hCKiGj')

# Initialize an empty list to store product data
product_data = []

# Regular expressions to extract brand and price
brand_pattern = r'^[A-Za-z\s]+'
price_pattern = r'₹[\d,]+'

# Loop through and extract the text of each element (product)
for elem in elems:
    product_info = elem.text
    
    # Extract Brand
    brand = re.search(brand_pattern, product_info)
    brand_name = brand.group(0) if brand else "Unknown"
    
    # Extract Price
    price = re.search(price_pattern, product_info)
    price_value = price.group(0) if price else "Not Available"
    
    # Extract Product Type (everything between brand and price)
    product_type = product_info.replace(brand_name, '').replace(price_value, '').strip()
    
    # If there's any extra info (like discounts), remove it
    product_type = re.sub(r'[\d,]+% off|Free delivery|Hot Deal|Only few left|Bank Offer|Lowest price since launch', '', product_type).strip()
    
    # Append the cleaned data
    product_data.append([brand_name, product_type, price_value])

# Close the browser after the operation
driver.quit()

# Create a pandas DataFrame to store the product data
df = pd.DataFrame(product_data, columns=["Brand", "Product Type", "Price"])

# Save the cleaned data to an Excel file
df.to_excel("flipkart_products_cleaned.xlsx", index=False)

# Print the cleaned dataframe to verify
print(df)

print("Data has been saved to 'flipkart_products_cleaned.xlsx'.")
```
